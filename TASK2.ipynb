```python
# sin(x/5) * exp(x/10) + 5 * exp(-x/2)

import matplotlib.pyplot as plt  
import numpy as np  

def func(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-x/2)

x = np.arange(1, 15, 0.01)
y = func(x)
plt.plot(x, y)
plt.show()
```

```python
y1 = func(1)
y2 = func(15)
print(y1)
print(y2)
```

```python
'''
w0 + w1 * 1 = 3.25222
w0 + w1 * 15 = 0.63522
'''
import scipy
from scipy import linalg
A = [[1,1],[1,15]]
b = [3.25222, 0.63522]
ans = linalg.solve(A, b)
print(ans)

```

```python
approx1 = 3.43914857 - 0.18692857 * x
plt.plot(x, y)
plt.plot(x, approx1)
plt.show()
```

```python
y3 = func(8)
print(y1, y3, y2)
```

```python
'''
w0 + w1 + w2 = 3.252216865271419
w0 + w1 * 8 + w2 * 64 = 2.316170159053224
w0 + w1 * 15 + w2 * 225 = 0.6352214195786656
'''
A2 = [[1,1, 1],[1,8, 64],[1,15,225]]
b2 = [3.25222, 2.31617, 0.63522]
ans2 = linalg.solve(A2, b2)
print(ans2)
```

```python
approx2 = 3.32513327 - 0.06531224 * x - 0.00760102 * x * x
plt.plot(x, y)
plt.plot(x, approx2)
plt.show()
```

```python
y4 = func(4)
y5 = func(10)
print(y1, y4, y5, y2)
```

```python
'''
w0 + w1 + w2 + w3 = 3.252216865271419
w0 + w1 * 4 + w2 * 16 + w3 * 64 = 1.7468459495903677
w0 + w1 * 10 + w2 * 100 + w3 * 1000 = 2.5054164070002463
w0 + w1 * 15 + w2 * 225 + w3 * 3375 = 0.6352214195786656
'''
A3 = [[1,1,1,1], [1,4,16,64], [1, 10, 100, 1000], [1, 15, 225, 3375]]
b3 = [3.252216865271419, 1.7468459495903677, 2.5054164070002463, 0.6352214195786656]
ans3 = linalg.solve(A3, b3)
print(ans3)
```
